```python
#Import Libraries
import pandas as pd
import numpy as np
import string, nltk, gensim
from nltk.tokenize import RegexpTokenizer
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
def read_text_file(filename):
    with open(filename) as f:
        lines = f.readlines()
    return lines
```

```python
file = read_text_file('Complete_Punjabi_SGGS.txt')
PBI_String= ' '.join(line for line in file)
review_data_list=PBI_String
filtered_words = [w for w in review_data_list.split(' ') if '॥' not in w]
```

```python
#create empty list
review_PBI = list()
indv_lines = file

for line in indv_lines:

    #create word tokens as well as remove puntuation in one go
    rem_tok_punc = RegexpTokenizer(r'\w+')

    tokens = rem_tok_punc.tokenize(line)


    #convert the words to lower case
    words = [w.lower() for w in tokens]

    #Invoke all the english stopwords
    stop_word_list = set()

    #Remove stop words
    words = [w for w in words if not w in stop_word_list]

    #Append words in the review_data_list list.
    review_PBI.append(words)
```

```python
#Train a Word2Vec model using Gensim
import gensim
Embedding_Dim = 100
#train word2vec model
model = gensim.models.Word2Vec(sentences = [filtered_words], vector_size = Embedding_Dim, workers = 4, min_count = 2)
#Vocabulary size
words = list(model.wv.key_to_index)
print('Here is the Vocabulary Size.. %d' % len(words))
```
